## Fibonacci (Recursive)

**Input:** Integer `n` (position in Fibonacci sequence).  
**Output:** The nth Fibonacci number.  

### Time Complexity:
- Best: O(n) (not applicable here, always recursive)  
- Average: O(2^n)  
- Worst: O(2^n)  

### Space Complexity:
- O(n) (due to recursion stack).  

### Suitability & Trade-offs:
- Easy to understand, but very slow for large n.  
- High recursion depth → stack overflow risk.  
- Not suitable for real-time applications.

```python
##Implementation
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

##Experimental Profiling
import time
from memory_profiler import memory_usage

n = 30

# Execution time
start = time.time()
fib_recursive(n)
end = time.time()
print(f"Recursive Fibonacci({n}) Time: {end-start:.6f} sec")

## Memory usage
def run_recursive():
    fib_recursive(n)
mem_usage = memory_usage(run_recursive)
print("Recursive Fibonacci Memory Used:", max(mem_usage)-min(mem_usage), "MB")

## Visualization
import matplotlib.pyplot as plt

sizes = [5, 10, 20, 25, 30]  # input sizes
times = []

for n in sizes:
    start = time.time()
    fib_recursive(n)
    end = time.time()
    times.append(end-start)

plt.plot(sizes, times, marker='o')
plt.xlabel("Input size (n)")
plt.ylabel("Execution time (seconds)")
plt.title("Recursive Fibonacci Performance")
plt.show()
```

### Discussion
- Execution time grows **exponentially** with n.  
- Memory usage increases with recursion depth.  
- Recursion depth risk for n > 1000.  
- Trade-off: Simple logic but impractical for large inputs.



### Iterative Fibonacci
**Input:** Integer `n`.  
**Output:** nth Fibonacci number.  

**Time Complexity:**  
- Best: O(n).  
- Average: O(n).  
- Worst: O(n).  

**Space Complexity:** O(1) (only a few variables).  

**Suitability & Trade-offs:**  
- Much faster than recursion.  
- Memory efficient.  
- Preferred for large inputs.




```python
#Iterative Implementation
def fib_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n+1):
        a, b = b, a+b
    return b

# Test
print("\n\nIterative Fibonacci up to 10 terms:")
for i in range(10):
    print(fib_iterative(i), end=" ")
##  Experimental Profiling
import time
from memory_profiler import memory_usage

n = 30  # Change for bigger tests

# Iterative time
start = time.time()
fib_iterative(n)
end = time.time()
print(f"Iterative Fibonacci({n}) Time: {end-start:.6f} sec")

##  Memory usage

def run_iterative():
    fib_iterative(n)

mem_iter = memory_usage(run_iterative)

print("Iterative Memory Used:", max(mem_iter)-min(mem_iter), "MB")
## Visualization
import matplotlib.pyplot as plt

sizes = [5, 10, 20, 25, 30]
times_recursive, times_iterative = [], []

for n in sizes:
    # Iterative
    start = time.time()
    fib_iterative(n)
    end = time.time()
    times_iterative.append(end-start)

plt.plot(sizes, times_iterative, marker='o', label="Iterative")
plt.xlabel("Input size (n)")
plt.ylabel("Execution time (seconds)")
plt.legend()
plt.show()
```

## Bubble Sort

**Input:** List of numbers.  
**Output:** Sorted list in ascending order.  

### Time Complexity:
- Best: O(n) (already sorted).  
- Average: O(n²).  
- Worst: O(n²).  

### Space Complexity:
- O(1) (in-place sorting).  

### Suitability & Trade-offs:
- Very simple, easy to implement.  
- Not suitable for large datasets.  
- Good for teaching basics of sorting.  


```python
## Implementation
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Test
print("Bubble Sort:", bubble_sort([64, 25, 12, 22, 11]))
```

## Selection Sort

**Input:** List of numbers.  
**Output:** Sorted list in ascending order.  

### Time Complexity:
- Best: O(n²).  
- Average: O(n²).  
- Worst: O(n²).  

### Space Complexity:
- O(1) (in-place).  

### Suitability & Trade-offs:
- Simpler but slower than efficient algorithms.  
- Performs fewer swaps compared to Bubble Sort.  
- Not practical for large datasets.  


```python
## Implementation
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Test
print("Selection Sort:", selection_sort([64, 25, 12, 22, 11]))
```

## Insertion Sort

**Input:** List of numbers.  
**Output:** Sorted list in ascending order.  

### Time Complexity:
- Best: O(n) (nearly sorted list).  
- Average: O(n²).  
- Worst: O(n²).  

### Space Complexity:
- O(1).  

### Suitability & Trade-offs:
- Good for small datasets or nearly sorted data.  
- Stable sort (preserves order of equal elements).  
- Not efficient for large datasets.  


```python
## Implementation
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

# Test
print("Insertion Sort:", insertion_sort([64, 25, 12, 22, 11]))
```

## Merge Sort

**Input:** List of numbers.  
**Output:** Sorted list in ascending order.  

### Time Complexity:
- Best: O(n log n).  
- Average: O(n log n).  
- Worst: O(n log n).  

### Space Complexity:
- O(n) (extra space for merging).  

### Suitability & Trade-offs:
- Very efficient, stable sorting algorithm.  
- Preferred for large datasets.  
- Uses recursion (risk of stack depth issues for extremely large n).  


```python
## Implementation
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

# Test
print("Merge Sort:", merge_sort([64, 25, 12, 22, 11]))
```

## Quick Sort

**Input:** List of numbers.  
**Output:** Sorted list in ascending order.  

### Time Complexity:
- Best: O(n log n).  
- Average: O(n log n).  
- Worst: O(n²) (if pivot choice is poor).  

### Space Complexity:
- O(log n) (recursion stack).  

### Suitability & Trade-offs:
- Very fast in practice (avg case).  
- In-place sorting.  
- Not stable (does not preserve equal element order).  


```python
## Implementation
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# Test
print("Quick Sort:", quick_sort([64, 25, 12, 22, 11]))
```

- **Trade-offs:**  
  - Use Bubble/Selection/Insertion for teaching or very small arrays.  
  - Use Merge/Quick for larger datasets.

## Binary Search

**Input:**  
- A sorted list of numbers.  
- A target value to search.  

**Output:**  
- Index of target element if found.  
- -1 if not found.  

### Time Complexity:
- Best: O(1) (if middle element is the target).  
- Average: O(log n).  
- Worst: O(log n).  

### Space Complexity:
- Iterative: O(1).  
- Recursive: O(log n) (stack frames).  

### Suitability & Trade-offs:
- Works only on **sorted arrays**.  
- Extremely efficient compared to linear search.  
- Recursive implementation risks stack overflow for very large arrays.  


```python
## Implementation
# Iterative Binary Search
def binary_search_iterative(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Recursive Binary Search
def binary_search_recursive(arr, low, high, target):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, mid+1, high, target)
    else:
        return binary_search_recursive(arr, low, mid-1, target)

# Test
arr = [10, 20, 30, 40, 50, 60, 70]
print("Iterative Search (40):", binary_search_iterative(arr, 40))
print("Recursive Search (40):", binary_search_recursive(arr, 0, len(arr)-1, 40))

## Profiling
import time
from memory_profiler import memory_usage

arr = list(range(1, 1000000))  # sorted array
target = 999999

# Iterative profiling
start = time.time()
binary_search_iterative(arr, target)
end = time.time()
print(f"Iterative Binary Search Time: {end-start:.6f} sec")

def run_iter():
    binary_search_iterative(arr, target)
mem_iter = memory_usage(run_iter)
print("Iterative Binary Search Memory:", max(mem_iter)-min(mem_iter), "MB")

# Recursive profiling
start = time.time()
binary_search_recursive(arr, 0, len(arr)-1, target)
end = time.time()
print(f"Recursive Binary Search Time: {end-start:.6f} sec")

def run_rec():
    binary_search_recursive(arr, 0, len(arr)-1, target)
mem_rec = memory_usage(run_rec)
print("Recursive Binary Search Memory:", max(mem_rec)-min(mem_rec), "MB")

## Visualization
sizes = [1000, 10000, 100000, 500000, 1000000]
times_iter, times_rec = [], []

for n in sizes:
    arr = list(range(1, n+1))
    target = n  # search last element

    # Iterative
    start = time.time()
    binary_search_iterative(arr, target)
    end = time.time()
    times_iter.append(end-start)

    # Recursive
    start = time.time()
    binary_search_recursive(arr, 0, len(arr)-1, target)
    end = time.time()
    times_rec.append(end-start)

# Plot Iterative
plt.plot(sizes, times_iter, marker='o', color='b')
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Iterative Binary Search Performance")
plt.show()

# Plot Recursive
plt.plot(sizes, times_rec, marker='o', color='r')
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Recursive Binary Search Performance")
plt.show()
```

### Discussion
- **Iterative Binary Search** is faster and more memory-efficient.  
- **Recursive Binary Search** is elegant but incurs O(log n) space due to recursion stack.  
- Both scale well with large datasets (logarithmic time).  
- Binary Search is only applicable to **sorted arrays**; unsorted arrays need sorting first (O(n log n)).
